```python
Math = ['Tom','Jerry','Spike','Tyke']
Political = ['Jerry','Spike','Butch','Nibbles']
Science = ['Spike','Tyke','Nibbles','Quacker']


M = set(Math)
P = set(Political)
S = set(Science)


all_three = M & P & S


two_subjects = (M & P) | (P & S) | (M & S)
two_subjects = two_subjects - all_three


all_names = M | P | S
one_subject = all_names - two_subjects - all_three

print("Only one subject =", list(one_subject))
print("Exactly two subjects =", list(two_subjects))
print("All subjects =", list(all_three))
```

monkey puzzle:odd /even usaing var i-"while"

```python
X = 21
Y = 5
Z = 3

time = 0
ht = 0
i = 1

while ht < X:


    if i % 2 == 1:
        ht = ht + Y
        time += 1


        if ht >= X:
            break


    else:
        ht = ht - Z
        time += 1

    i += 1

print(time)
```

fizz buzz- condition priority

```python
n = int(input("Enter n: "))
answer = []

for i in range(1, n + 1):
    if i % 3 == 0 and i % 5 == 0:
        answer.append("FizzBuzz")
    elif i % 3 == 0:
        answer.append("Fizz")
    elif i % 5 == 0:
        answer.append("Buzz")
    else:
        answer.append(str(i))

print(answer)
```
