```python
l=int(input("enter legs: "))
h=int(input("enter heads: "))
flag=False
for cow in range(0,h+1):
    hens=h-cow
    cal=cow*4+hens*2
    if cal==l:
        flag=True
        break
if flag:
    print("cows: ",cow)
    print("hens: ",hens)
else:
    print("no solution")
         
```
