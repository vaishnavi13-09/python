```python
import numpy as np
import matplotlib.pyplot as plt
overs=np.arange(1,11)
scores=[2,4,7,1,12,8,5,3,9,10]
plt.bar(overs,scores,color="green")
plt.title("INDIA SCORE BASED ON OVERS")
plt.xlabel("overs from 10")
plt.ylabel("runs by overs")
plt.xticks(overs)
plt.grid(axis="y",linestyle="-")
plt.show()
```

percentile tells you how a value compares with other values in a dataset.

```python
import numpy as np
a=[10,20,30,40,50,60]
res=np.percentile(a,50)
print(res)
```

IMPLEMENTATION OF STACK:

```python
class stack:
  def __init__ (self):
    self.l=[]
  def push(self,a):
    self.l.append(a)
  def pop(self):
    if not self.l:
        print("Stack is empty!")
        return
    self.l.pop()
  def display(self):
    if not self.l:
        print("Stack is empty")
        return
    b="->".join(map(str,self.l))
    print(b)
  def peek(self):
    if not self.l:
        print("Stack is empty!")
        return
    print(self.l[-1])
s=stack()
s.push(100)
s.push(30)
s.push(56)
s.display()
s.pop()
s.display()
s.peek()
```
