valid perfect squares

```python
class Solution:
    def isPerfectSquare(self, num: int) -> bool:
        a = num ** 0.5
        if a % 1 == 0:
            return True
        else:
            return False
```

password-Real time application

```python
a = input("enter your password: ")
up = 0
sm = 0
sp = 0
dg = 0
if len(a) > 7:
    for i in a:
        if i.isupper():
            up += 1
        elif i.islower():
            sm += 1
        elif i.isdigit():
            dg += 1
        else:
            sp += 1
    if up > 0 and sm > 0 and dg > 0 and sp > 0:
        print("password is strong")
    else:
        print("password is weak")
else:
    print("weak due to less char")
```

Valid Palindrome

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        a = " "
        for ch in s:
            if ch.isalnum():
                a+= ch.lower()

        return a == a[::-1]
```
